```python
#rename ambiguous or not standardized collumn names
df_customers_unambiguous = df_customers.rename(columns={'First Name':'first_name','Surnam':'surname','Gender':'gender','STATE':'state','Age':'age','n_dependants':'number_dependants','fam_status':'family_status'})
```

### Find Mixed Value collumns

```python
#print columns with mixed data types
for col in df_customers_unambiguous.columns.tolist():
  weird = (df_customers_unambiguous[[col]].map(type) != df_customers_unambiguous[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers_unambiguous[weird]) > 0:
    print (col)
```

```python
#change type of first_name column to string/object
df_customers_unambiguous['first_name']= df_customers_unambiguous['first_name'].astype('str')
```

```python
#Recheck and print columns with mixed data types
for col in df_customers_unambiguous.columns.tolist():
  weird = (df_customers_unambiguous[[col]].map(type) != df_customers_unambiguous[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers_unambiguous[weird]) > 0:
    print (col)
```

### Find Missing Values
