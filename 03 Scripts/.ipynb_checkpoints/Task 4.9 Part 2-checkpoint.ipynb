## Task 5. Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day. (Hint: To check this, you need to use an accurate sample for your line chart!)

```python
#Create a random seed for reproducibility
np.random.seed(4)
#create a new list, dev. Stores boolean values—output of True or False for 32,404,859 random numbers wether it’s less than or equal to 0.7
dev = np.random.rand(len(final_df)) <= 0.7
```

```python
#Store 70% of our df's values in big
big = final_df[dev]
#Store the remaining 30% in small (~ negates the False to become True)
small = final_df[~dev]
```

```python
#Creation of line chart of H.O.D. versus prices using small subsect
sns.lineplot(x='order_hour_of_day',y='prices',data=small)
plt.title('Avg Price of items bought throughout the day')
plt.xlabel('Hour of the day')
plt.ylabel('Avg Price')
```

```python
#Creation of line chart of H.O.D. versus prices using big subsect
line_hod_vs_prices_big = sns.lineplot(x='order_hour_of_day',y='prices',data=big)
```

We can see that both the big and small sample plots are relatively comparable, so we can assume that the small sample has a similar behaviour to the whole dataset. However, what I don't understand exactly is what is actually being displayed in the charts. Is it the mean price for each hour of the day, with the error in blue? Does this mean I do not need to group the data to analyze it? To better understand this, I aimed to group the data in different ways and analyze it in more detail.

```python
#creating a new dataframe with price information grouped by hour of day and aggregated first by mean and then sum
df_expenditure_mean = final_df.groupby('order_hour_of_day')['prices'].mean()
df_expenditure_sum = final_df.groupby('order_hour_of_day')['prices'].sum()
```

```python
# Resetting the index of the dataframe to have both columns available for analysis
df_expenditure_mean= pd.DataFrame(df_expenditure_mean).reset_index()
df_expenditure_sum= pd.DataFrame(df_expenditure_sum).reset_index()
```

```python
#lineplot of the mean price grouped by hour of the day
line_meanprice_vs_hod = sns.lineplot(data=df_expenditure_mean, x='order_hour_of_day',y='prices')
```

```python
#lineplot of the sum of the price grouped by hour of the day
line_totalprice_vs_hod = sns.lineplot(data=df_expenditure_sum, x='order_hour_of_day',y='prices')
```

As we can see, the initial plot using seaborn used the mean to aggregate the data, as that graph is the more similar to the meanp-price-aggregated one. Interestingly, the total-price-aggregated graph shows a similar behaviour to the number of orders per hour of day, which is to be expected. As the number of orders increases, the total expenditure is also expected to increase.

## Task 7. You’ll also need to explore whether there’s a connection between age and spending power (income). To visualize this relationship, create a scatterplot using the sns.scatterplot() function.

```python
#generating new DataFrame with the median(due to common income outliers) income grouped by age
df_income_by_age = final_df.groupby('age')['income'].median()
```

```python
#Resetting index to get both dimensions as collumns in dataframe
df_income_by_age = pd.DataFrame(df_income_by_age).reset_index()
```

```python
#Checking results
df_income_by_age.head(50)
```

```python
#Plotting scatterplot of median income vs age
scatter_age_vs_income = sns.scatterplot(data=df_income_by_age,x='age',y='income')
plt.title('Median income for different ages')
plt.xlabel('Age')
plt.ylabel('Median Income')
plt.savefig(os.path.join(path,'04 Analysis','Visualizations','Median_income_for_diff_ages.png'))
```

There seems to be a relatively stable median income up till 41 years, at which point there is a ~70% increase in median income, which stays stable untill the oldest population.
